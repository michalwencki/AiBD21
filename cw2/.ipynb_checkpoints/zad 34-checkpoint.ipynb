```python
## Michał Wencki
## Zdefiniuj poniższą funkcję i sporządź jej wykres dla argumentów z danego przedziału:
## $f(x)=x^2+5$

##     1.$x>-1 oraz x<1$
##     2.$x>-6 oraz x<6$
##     3.$x>0 oraz x<5$
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def calculate_function3(x: float)-> float:
    return x**2+5
```

```python
t1 = np.arange(-1.0, 1.0, 0.01)
t2 = np.arange(-6.0, 6.0, 0.01)
t3 = np.arange(0, 5.0, 0.01)
fig, ax = plt.subplots()
ax.plot(t2, calculate_function3(t2), label='Funkcja x^2+5 dla x:<-6;6>')
ax.plot(t1, calculate_function3(t1), label='Funkcja x^2+5 dla x:<-1;1>')
ax.plot(t3, calculate_function3(t3), label='Funkcja x^2+5 dla x:<0;5>')
ax.set(xlabel='x', ylabel='x^2+5',
       title='3 funkcje x^2+5 na jednym wykresie')
ax.grid()
legend = ax.legend(fancybox=True, shadow=True)
```

```python
t1 = np.arange(-1.0, 1.0, 0.01)
t2 = np.arange(-6.0, 6.0, 0.01)
t3 = np.arange(0, 5.0, 0.01)
fig, (ax1, ax2, ax3) = plt.subplots(3, 1)
fig.suptitle('3 funkcje x^2+5 na 3 różnych wykresach')
ax1.plot(t2, calculate_function3(t2), label='Funkcja x^2+5 dla x:<-6;6>')
ax2.plot(t1, calculate_function3(t1), label='Funkcja x^2+5 dla x:<-1;1>')
ax3.plot(t3, calculate_function3(t3), label='Funkcja x^2+5 dla x:<0;5>')
ax1.set(xlabel='x', ylabel='x^2+5')
ax2.set(xlabel='x', ylabel='x^2+5')
ax3.set(xlabel='x', ylabel='x^2+5')
ax1.grid()
ax2.grid()
ax3.grid()
legend = ax.legend(fancybox=True, shadow=True)
```

```python
from dataclasses import make_dataclass
import pandas
Person = make_dataclass("Person", [("name", str), ("surname", str), ("age", int), ("sex", str)])
example_dataframe = pandas.DataFrame([Person("Gary", "Kowalski", 16, "Male"), Person("Fiodor", "Raskolnikow", 49, "Male"),
                                    Person("Julia", "Capuleti", 12, "Female"), Person("Franek", "Kimono", 69, "Male"),
                                    Person("Michele", "Mouton", 70, "Female")])
pandas.DataFrame.info(example_dataframe)
```
